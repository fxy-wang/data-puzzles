# Problem summary


Can we extract the birth date of that person from that set of data?

*Some starter code:*



```
import pandas as pd
import matplotlib.pyplot as plt

#Â Load data
dd_df = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/regression/chal1/dd.csv')
mm_df = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/regression/chal1/mm.csv')
yyyy_df = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/regression/chal1/yyyy.csv')
```

```
# Plot the DD data
plt.scatter(dd_df.x, dd_df.y)
```

```
# Plot the MM data
plt.scatter(mm_df.x, mm_df.y)
```

```
# Plot the YYYY data
plt.scatter(yyyy_df.x, yyyy_df.y)
```

# Your turn

Find the birth date 'encrypted' in these datasets!

```
## Date Extraction
# Use linear regression

from sklearn.linear_model import LinearRegression

day = LinearRegression()

ddx = dd_df.x.values
ddy = dd_df.y.values

day.fit(ddx.reshape(-1,1),ddy.reshape(-1,1))
```

```
print(day.coef_)
print(day.intercept_)
```

```
## Month Extraction

month = LinearRegression()

mmx = mm_df.x.values
mmy = mm_df.y.values

month.fit(mmx.reshape(-1,1),mmy.reshape(-1,1))

```

```
print(month.coef_)
print(month.intercept_)
```

```
## Year Extraction
# Use polynomial fit
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3,include_bias=False)

yyx = yyyy_df.x.values
yyy = yyyy_df.y.values

x_poly = poly.fit_transform(yyx.reshape(-1,1))
y_poly = poly.fit_transform(yyy.reshape(-1,1))
```

```
year = LinearRegression()

year.fit(x_poly,yyy.reshape(-1,1))

```

```
print(year.coef_)
print(year.intercept_)
```
