# Problem summary


The card says:
*If lost please send back to the most predictive features of...*



```
import pandas as pd

# Load data
table = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/regression/chal1/reg_p2.csv')
```

```
# Print the table
table
```

Can we find the most predictive features of this table to find the proper address to return the lost wallet?

# Your turn


```
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler

X = table[table.columns[0:-1]]
Y = table[table.columns[-1]]


scaler = StandardScaler().fit(X)
Xscaled = scaler.transform(X)
```

```
sel_ = SelectFromModel(Lasso(alpha=50,max_iter=1000),threshold=10)
sel_.fit(Xscaled,Y.values)
```

```
selected = X.columns[(sel_.get_support())]
```
